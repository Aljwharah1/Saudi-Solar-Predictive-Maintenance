```python
import pandas as pd 
import numpy as np  


get_df =pd.read_csv("../data/Plant_1_Weather_Sensor_Data.csv")
weather_df = pd.read_csv("../data/Plant_1_Weather_Sensor_Data.csv")


get_df['DATE_TIME'] = pd.to_datetime(get_df["DATE_TIME"])
weather_df['DATE_TIME']=pd.to_datetime(weather_df["DATE_TIME"])

df = pd.merge(get_df, weather_df , on="DATE_TIME" ,how = 'inner')

print(df.shape)
df.head()
```

```python
columns_keep = [
    'DATE_TIME', 
    'PLANT_ID_x', 
    'AMBIENT_TEMPERATURE_x', 
    'MODULE_TEMPERATURE_x', 
    'IRRADIATION_x'
]

copy_col =df[columns_keep].copy()

copy_col.columns= [
    'date_time', 
    'plant_id', 
    'ambient_temp', 
    'module_temp', 
    'irradiation'
]
```

```python
np.random.seed(42)
copy_col['dust_level']=np.random.uniform(0.1,0.9,len(copy_col))
```

```python
def maintenace_logic(row):
    if row["dust_level"]>0.7 or row['module_temp']>60:
        return 1 
    else: 
        return 0 
copy_col['needs_maintenance']=copy_col.apply(maintenace_logic,axis=1)
print("done to add")
    
```
